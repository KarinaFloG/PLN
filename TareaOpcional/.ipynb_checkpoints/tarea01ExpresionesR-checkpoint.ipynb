```python
#Importa paquetes
import nltk
import re
#NLTK contiene corpus en español
from nltk.corpus import cess_esp
```

```python
#Palabras ejemplo
singulares = ['amar', 'comer', 'vivir', 'hablar', 'beber']
primer_singular = []
segundo_singular = []
tercer_singular = []
primera_persona_p = []
segunda_persona_p = []
tercera_persona_p = []

for sg in singulares:
  #Se busca las palabras que terminan en consonante o 'i'
    if re.match(r'[^ ]+ar\b',sg):
        #Se concatena 'o'
        ps = re.sub(r'ar\b',r'o', sg)
        primer_singular.append(ps)
        #Se concatena 'ar'
        ss = re.sub(r'ar\b',r'as', sg)
        segundo_singular.append(ss)
        #Se concatena 'a'
        ts = re.sub(r'ar\b',r'a', sg)
        tercer_singular.append(ts)
        #Se concatena 'amos'
        ppp = re.sub(r'ar\b',r'amos', sg)
        primera_persona_p.append(ppp)
        #Se concatena 'áis'
        spp = re.sub(r'ar\b',r'áis', sg)
        segunda_persona_p.append(spp)
        #Se concatena 'an'
        tpp = re.sub(r'ar\b',r'an', sg)
        tercera_persona_p.append(tpp)
    elif re.match(r'[^ ]+er\b',sg):
        #Se concatena 'o'
        ps = re.sub(r'er\b',r'o', sg)
        primer_singular.append(ps)
        #Se concatena 'er'
        ss = re.sub(r'er\b',r'er', sg)
        segundo_singular.append(ss)
        #Se concatena 'e'
        ts = re.sub(r'er\b',r'e', sg)
        tercer_singular.append(ts)
        #Se concatena 'emos'
        ppp = re.sub(r'er\b',r'emos', sg)
        primera_persona_p.append(ppp)
        #Se concatena 'éis'
        spp = re.sub(r'er\b',r'éis', sg)
        segunda_persona_p.append(spp)
        #Se concatena 'en'
        tpp = re.sub(r'er\b',r'en', sg)
        tercera_persona_p.append(tpp)
    elif re.match(r'[^ ]+ir\b',sg):
        #Se concatena 'i'
        ps = re.sub(r'ir\b',r'i', sg)
        primer_singular.append(ps)
        #Se concatena 'es'
        ss = re.sub(r'ir\b',r'es', sg)
        segundo_singular.append(ss)
        #Se concatena 'e'
        ts = re.sub(r'ir\b',r'e', sg)
        tercer_singular.append(ts)
        #Se concatena 'imos'
        ppp = re.sub(r'ir\b',r'imos', sg)
        primera_persona_p.append(ppp)
        #Se concatena 'áis'
        spp = re.sub(r'ir\b',r'ís', sg)
        segunda_persona_p.append(spp)
        #Se concatena 'amamos'
        tpp = re.sub(r'ir\b',r'en', sg)
        tercera_persona_p.append(tpp)
    else:
        print(sg+'s')

        
      
    

print("Verbos en singular")
print(singulares)

print("Verbos primera persona del singular")
print(primer_singular)

print("Verbos segunda persona del singular")
print(segundo_singular)

print("Verbos tercera persona del singular")
print(tercer_singular)

print("Verbos primera persona del plural")
print(primera_persona_p)

print("Verbos segunda persona del plural")
print(segunda_persona_p)

print("Verbos tercera persona del plural")
print(tercera_persona_p)
```
